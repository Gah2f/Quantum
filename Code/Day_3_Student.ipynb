```python
# Check the |+> state

qubit = cirq.NamedQubit('q0')

circuit_plus_no_measure = cirq.Circuit()
circuit_plus_no_measure.append(cirq.H(qubit))

print(circuit_plus_no_measure)
```

```python
# State vector and Dirac notation for |+>

vec_plus = cirq.final_state_vector(
    circuit_plus_no_measure,
    ignore_terminal_measurements = True
)

print(vec_plus)
print(cirq.dirac_notation(vec_plus))
```

```python
cirq_web.bloch_sphere.BlochSphere(state_vector = vec_plus)
```

```python
# Check the |-> state

qubit = cirq.NamedQubit('q0')

circuit_minus_no_measure = cirq.Circuit()
circuit_minus_no_measure.append(cirq.H(qubit))
circuit_minus_no_measure.append(cirq.Z(qubit))

print(circuit_minus_no_measure)
```

```python
# State vector and Dirac notation for |->

vec_minus = cirq.final_state_vector(
    circuit_minus_no_measure,
    ignore_terminal_measurements = True
)

print(vec_minus)
print(cirq.dirac_notation(vec_minus))
```

```python
# Bloch sphere for |->

cirq_web.bloch_sphere.BlochSphere(state_vector = vec_minus)
```
